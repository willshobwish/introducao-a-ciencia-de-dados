```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy
import numpy as np
from sklearn.preprocessing import LabelEncoder, StandardScaler, Normalizer, OneHotEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression, LinearRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, log_loss, classification_report
```

```python
# Show all rows
pd.set_option('display.max_rows', None)

# Show all columns
pd.set_option('display.max_columns', None)

# Optional: Set wider column display for better visibility
pd.set_option('display.width', 1000)
```

```python
df = pd.read_csv("predict_students_dropout_and_academic_success.csv",delimiter=";")
df_encoded = df.copy()
```

```python
print(df_encoded.columns)
```

```python
df_encoded.head()
```

É necessário fazer encoding dos dados (e.g. curso, qualificação, ocupação e dos pais)

Portanto será feito inicialmente uma analise das features que são categoricas e aplicar o encoding

```python
fields = [
    "Course",
    "Mother's qualification",
    "Father's qualification",
    "Mother's occupation",
    "Father's occupation",
    "Application mode",
    "Target",
    "Nacionality",
    "Application mode",
    "Application order"
]

for columns in fields:
    le = LabelEncoder()
    df_encoded[columns]=le.fit_transform(df[columns])
```

```python
display(df.head(5))
display(df_encoded.head(5))
```

Normalizaremos os dados que possam apresentar distancias grandes

```python
for columns in df_encoded:
    print(f"{columns}: {len(df_encoded[columns].unique())}")
```

```python
df_normalized = df_encoded.copy()
for i in ["Previous qualification (grade)",
    "Admission grade",
    "Age at enrollment",
    "Curricular units 1st sem (credited)",
    "Curricular units 1st sem (enrolled)",
    "Curricular units 1st sem (evaluations)",
    "Curricular units 1st sem (approved)",
    "Curricular units 1st sem (grade)",
    "Curricular units 1st sem (without evaluations)",
    "Curricular units 2nd sem (credited)",
    "Curricular units 2nd sem (enrolled)",
    "Curricular units 2nd sem (evaluations)",
    "Curricular units 2nd sem (approved)",
    "Curricular units 2nd sem (grade)",
    "Curricular units 2nd sem (without evaluations)",
    "Unemployment rate",
    "Inflation rate",
    "GDP"]:
    normalizer = StandardScaler()
    df_normalized[i] = normalizer.fit_transform(df_encoded[[i]])
```

```python
display(df_normalized.head(5))
```

```python
# df_oneHot = df_encoded.copy()
df_oneHot = pd.DataFrame()
for i in ["Course",
    "Mother's qualification",
    "Father's qualification",
    "Mother's occupation",
    "Father's occupation",
    "Application mode",
    "Target",
    "Nacionality",
    "Application order"]:
    
    ohe = OneHotEncoder(sparse_output=False)

    encoded = ohe.fit_transform(df_encoded[[i]])
    encoded_df = pd.DataFrame(encoded, columns=ohe.get_feature_names_out([i]))
    df_oneHot = pd.concat([df_oneHot, encoded_df], axis=1)
    # df_oneHot = df_oneHot.drop(columns=[i])
df_oneHot = pd.concat([df_oneHot, df_encoded["Target"]], axis=1)

display(df_oneHot.head(5))
```

Aplicando regressão logistica

```python
max_iter=10000
```

```python
regressionOneHot = LogisticRegression(max_iter=max_iter)

display(df_oneHot.iloc[:, :-1].head(5))
display(df_oneHot.iloc[:, -1].head(5))
x_train, x_test, y_train, y_test = train_test_split(df_oneHot.iloc[:, :-1], df_oneHot.iloc[:, -1],test_size=0.2,random_state=42)
regressionOneHot.fit(X=x_train,y=y_train)
y_pred = regressionOneHot.predict(X=x_test)
y_pred_proba = regressionOneHot.predict_proba(X=x_test)
# Metrics
acc = accuracy_score(y_test, y_pred)
prec = precision_score(y_test, y_pred,average="weighted")
rec = recall_score(y_test, y_pred,average="weighted")
f1 = f1_score(y_test, y_pred,average="weighted")
loss = log_loss(y_test, y_pred_proba)

# Output
print(f"Accuracy:  {acc:.4f}")
print(f"Precision: {prec:.4f}")
print(f"Recall:    {rec:.4f}")
print(f"F1 Score:  {f1:.4f}")
print(f"Log Loss:  {loss:.4f}")
```

```python
linearRegressionOneHot = LinearRegression()

display(df_oneHot.iloc[:, :-1].head(5))
display(df_oneHot.iloc[:, -1].head(5))
x_train, x_test, y_train, y_test = train_test_split(df_oneHot.iloc[:, :-1], df_oneHot.iloc[:, -1],test_size=0.2,random_state=42)
linearRegressionOneHot.fit(X=x_train,y=y_train)
y_pred = linearRegressionOneHot.predict(X=x_test)
# y_pred_proba = linearRegressionOneHot.predict_proba(X=x_test)
# Metrics
# Predict and round to get class labels
y_pred = np.round(y_pred).astype(int)

# Ensure predictions are within valid class labels (optional but safe)
y_pred = np.clip(y_pred, y_test.min(), y_test.max())

# Now you can compute classification metrics
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Precision:", precision_score(y_test, y_pred, average='macro'))
print("Recall:", recall_score(y_test, y_pred, average='macro'))
print("F1 Score:", f1_score(y_test, y_pred, average='macro'))
# print(f"Log Loss:  {loss:.4f}")
```
